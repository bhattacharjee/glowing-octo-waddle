```python
df = pd.read_csv('sorted_ratio.csv.gz').fillna(0)
df
```

```python
def is_country(x):
    if "countries" in x.lower() or "country" in x.lower():
        return False
    if "south africa" == x.lower():
        return True
    if "total" in x.lower():
        return False
    if "Latin America & Caribbean" == x:
        return False
    if x.lower() in ["world", "middle income", "ibrd only", "oecd", "euro area"]:
        return False
    for region in ["asia", "africa", "europe", "middle"]:
        if region in x.lower():
            return False
    for keyword in ['IDA', 'Fragile', 'demographic', 'World', 'IBRD', 'Income', 'income']:
        if keyword in x:
            return False
    return True
```

```python
df[df['country'].map(is_country) == False].reset_index(drop=True).rename(columns={"country": "country type"})
```

```python
df[df['country'].map(is_country) == True].head(40).reset_index(drop=True)
```

```python
countrydf = df[df['country'].map(is_country) == True].reset_index(drop=True)
```

```python
def interesting_country(x):
    if "united states" in x.lower():
        return True
    for c in ['france', 'germany', 'denmark', 'netherland', 'belgium', 'austria', 'sweden', 'switzerland', 'ireland', 'canada', 'united kingdom', 'italy', 'spain', 'portugal']:
        if c in x.lower():
            return True
    return False

countrydf[countrydf['country'].map(interesting_country)]
```
