### Using a different Ratio

$ ratio1 = \frac {Refugees} {Total\ GDP\ (PPP)} $

$ ratio1 = \frac {Refugees} {Population} $

The final ratio is the harmonic mean of the two ratios:

$ final\ measure = \frac {2 \times ratio1 \times ratio2} {ratio1 + ratio2} $

```python
!ls
import pandas as pd
```

```python
df = pd.read_csv('combined.csv.gz').fillna(0)
print(df.columns)

def normalize(df):
    return df
    df = (df - df.min()) / (df.max() - df.min())
    return df
    
df['ratio1'] = df['refugees'] / df['gdp-ppp']
df['ratio2'] = df['refugees'] / df['population']
df = df[(df['ratio1'] != float('inf')) & (df['ratio2'] != float('inf'))]
df['ratio1'] = normalize(df['ratio1'])
df['ratio2'] = normalize(df['ratio2'])
df['ratio'] = 2 * df['ratio1'] * df['ratio2'] / (df['ratio1'] + df['ratio2']) 
df = df.sort_values(by='ratio', ascending=False).reset_index(drop=True)
df = df[['country', 'ratio']]
df['ratio'] = df['ratio'] * 1.0e2
```

```python
def is_country(x):
    if "countries" in x.lower() or "country" in x.lower():
        return False
    if "south africa" == x.lower():
        return True
    if "total" in x.lower():
        return False
    if "Latin America & Caribbean" == x:
        return False
    if x.lower() in ["world", "middle income", "ibrd only", "oecd", "euro area"]:
        return False
    for region in ["asia", "africa", "europe", "middle"]:
        if region in x.lower():
            return False
    for keyword in ['IDA', 'Fragile', 'demographic', 'World', 'IBRD', 'Income', 'income']:
        if keyword in x:
            return False
    return True
```

```python
df[df['country'].map(is_country) == False].reset_index(drop=True).rename(columns={"country": "country type"})
```
